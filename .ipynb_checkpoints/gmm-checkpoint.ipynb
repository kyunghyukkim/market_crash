### Log likelihood 
We will keep looping through EM update steps until the log likehood ceases to increase at a certain rate.

```python
def log_sum_exp(Z):
    """ Compute log(\sum_i exp(Z_i)) for some array Z."""
    return np.max(Z) + np.log(np.sum(np.exp(Z - np.max(Z))))

def loglikelihood(data, weights, means, covs):
    """ Compute the loglikelihood of the data for a Gaussian mixture model with the given 
    parameters. """
    num_clusters = len(means)
    num_dim = len(data[0])
    
    ll = 0
    for d in data:
        
        Z = np.zeros(num_clusters)
        for k in range(num_clusters):
            
            # Compute (x-mu)^T * Sigma^{-1} * (x-mu)
            delta = np.array(d) - means[k]
            exponent_term = np.dot(delta.T, np.dot(np.linalg.inv(covs[k]), delta))
            
            # Compute loglikelihood contribution for this data point and this cluster
            Z[k] += np.log(weights[k])
            Z[k] -= 1/2. * (num_dim * np.log(2*np.pi) + np.log(np.linalg.det(covs[k])) 
                            + exponent_term)
            
        # Increment loglikelihood contribution of this data point across all clusters
        ll += log_sum_exp(Z)
        
    return ll
```

### E-step: assign cluster responsibilities, given current parameters

The first step in the EM algorithm is to compute cluster responsibilities. Let $r_{ik}$ denote the responsibility of cluster $k$ for data point $i$. Note that cluster responsibilities are fractional parts: Cluster responsibilities for a single data point $i$ should sum to 1.
$$
r_{i1} + r_{i2} + \ldots + r_{iK} = 1
$$

To figure how much a cluster is responsible for a given data point, we compute the likelihood of the data point under the  particular cluster assignment, multiplied by the weight of the cluster. For data point $i$ and cluster $k$, this quantity is
$$
r_{ik} \propto \pi_k N(x_i | \mu_k, \Sigma_k)
$$
where $N(x_i | \mu_k, \Sigma_k)$ is the Gaussian distribution for cluster $k$ (with mean $\mu_k$ and covariance $\Sigma_k$).

We used $\propto$ because the quantity $N(x_i | \mu_k, \Sigma_k)$ is not yet the responsibility we want. To ensure that all responsibilities over each data point add up to 1, we add the normalization constant in the denominator:
$$
r_{ik} = \frac{\pi_k N(x_i | \mu_k, \Sigma_k)}{\sum_{k=1}^{K} \pi_k N(x_i | \mu_k, \Sigma_k)}.
$$

**Drawing from a Gaussian distribution.** SciPy provides a convenient function [multivariate_normal.pdf](http://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.multivariate_normal.html) that computes the likelihood of seeing a data point in a multivariate Gaussian distribution. 

```python
def compute_responsibilities(data, weights, means, covariances):
    '''E-step: compute responsibilities, given the current parameters'''
    num_data = len(data)
    num_clusters = len(means)
    resp = np.zeros((num_data, num_clusters))
    
    # Update resp matrix so that resp[i,k] is the responsibility of cluster k for data point i.
    for i in range(num_data):
        for k in range(num_clusters):
            resp[i, k] = np.float32(weights[k] * multivariate_normal.pdf(data[i,:], means[k], 
                                                                         covariances[k]))
            if np.isnan(resp[i,k]):
                print "np.nan exists in resp!"
                exit(1)
    # Add up responsibilities over each data point and normalize
    row_sums = resp.sum(axis=1)[:, np.newaxis]
    resp = resp / row_sums
    return resp
```

### M-step: Update parameters, given current cluster responsibilities

Once the cluster responsibilities are computed, we update the parameters (weights, means, and covariances) associated with the clusters.

**Computing soft counts**. Before updating the parameters, we first compute what is known as "soft counts". The soft count of a cluster is the sum of all cluster responsibilities for that cluster:
$$
N^{\text{soft}}_k = r_{1k} + r_{2k} + \ldots + r_{Nk} = \sum_{i=1}^{N} r_{ik}
$$

where we loop over data points. Note that, unlike k-means, we must loop over every single data point in the dataset. This is because all clusters are represented in all data points, to a varying degree.

```python
def compute_soft_counts(resp):
    # Compute the total responsibility assigned to each cluster, which will be useful when 
    # implementing M-steps below. In the lectures this is called N^{soft}
    counts = np.nansum(resp, axis=0)
    return counts
```

**Updating weights.** The cluster weights show us how much each cluster is represented over all data points. The weight of cluster $k$ is given by the ratio of the soft count $N^{\text{soft}}_{k}$ to the total number of data points $N$:
$$
\hat{\pi}_k = \frac{N^{\text{soft}}_{k}}{N}
$$

```python
def compute_weights(counts):
    num_clusters = len(counts)
    weights = [0.] * num_clusters
    total_counts = np.sum(counts)
    
    for k in range(num_clusters):
        # Update the weight for cluster k using the M-step update rule for the cluster weight, 
        \hat{\pi}_k.
        weights[k] = counts[k]/total_counts

    return weights
```

**Updating means**. The mean of each cluster is set to the [weighted average](https://en.wikipedia.org/wiki/Weighted_arithmetic_mean) of all data points, weighted by the cluster responsibilities:
$$
\hat{\mu}_k = \frac{1}{N_k^{\text{soft}}} \sum_{i=1}^N r_{ik}x_i
$$

```python
def compute_means(data, resp, counts):
    num_clusters = len(counts)
    num_data = len(data)
    means = [np.zeros(len(data[0]))] * num_clusters
    
    for k in range(num_clusters):
        # Update means for cluster k using the M-step update rule for the mean variables.
        # This will assign the variable means[k] to be our estimate for \hat{\mu}_k.
        weighted_sum = 0.
        for i in range(num_data):
            if np.isnan(resp[i,k]):
                resp[i,k]=0
            weighted_sum += resp[i,k] * data[i]
        means[k] = weighted_sum/counts[k]

    return means
```

**Updating covariances**.  The covariance of each cluster is set to the weighted average of all [outer products](https://people.duke.edu/~ccc14/sta-663/LinearAlgebraReview.html), weighted by the cluster responsibilities:
$$
\hat{\Sigma}_k = \frac{1}{N^{\text{soft}}_k}\sum_{i=1}^N r_{ik} (x_i - \hat{\mu}_k)(x_i - \hat{\mu}_k)^T
$$

The "outer product" in this context refers to the matrix product
$$
(x_i - \hat{\mu}_k)(x_i - \hat{\mu}_k)^T.
$$
Letting $(x_i - \hat{\mu}_k)$ to be $d \times 1$ column vector, this product is a $d \times d$ matrix. Taking the weighted average of all outer products gives us the covariance matrix, which is also $d \times d$.

```python
def compute_covariances(data, resp, counts, means):
    num_clusters = len(counts)
    num_dim = len(data[0])
    num_data = len(data)
    covariances = [np.zeros((num_dim,num_dim))] * num_clusters
    
    for k in range(num_clusters):
        # Update covariances for cluster k using the M-step update rule for covariance 
        # variables.
        # This will assign the variable covariances[k] to be the estimate for \hat{\Sigma}_k.
        weighted_sum = np.zeros((num_dim, num_dim))
        for i in range(num_data):
            diff = data[i] - means[k]
            weighted_sum += resp[i,k] * diff * diff[:, np.newaxis]
            
        covariances[k] = weighted_sum/counts[k]

    return covariances
```

### The EM algorithm

Let us write a function that takes initial parameter estimates and runs EM.

```python
def EM(data, init_means, init_covariances, init_weights, maxiter=1000, thresh=1e-4):
    
    # Make copies of initial parameters, which we will update during each iteration
    means = init_means[:]
    covariances = init_covariances[:]
    weights = init_weights[:]
    
    # Infer dimensions of dataset and the number of clusters
    num_data = len(data)
    num_dim = len(data[0])
    num_clusters = len(means)
    
    # Initialize some useful variables
    resp = np.zeros((num_data, num_clusters))
    ll = loglikelihood(data, weights, means, covariances)
    ll_trace = [ll]
    
    for it in range(maxiter):
        if it % 5 == 0:
            print("Iteration %s" % it)
        
        # E-step: compute responsibilities
        covariances = covariances
        
        resp = compute_responsibilities(data, weights, means, covariances)
        
        # M-step
        # Compute the total responsibility assigned to each cluster, which will be useful when 
        # implementing M-steps below. In the lectures this is called N^{soft}
        counts = compute_soft_counts(resp)
        # Update the weight for cluster k using the M-step update rule for the cluster weight, 
        # \hat{\pi}_k.
        weights = compute_weights(counts)
        # Update means for cluster k using the M-step update rule for the mean variables.
        # This will assign the variable means[k] to be our estimate for \hat{\mu}_k.
        means = compute_means(data, resp, counts)
        # Update covariances for cluster k using the M-step update rule for covariance variables.
        # This will assign the variable covariances[k] to be the estimate for \hat{\Sigma}_k.
        covariances = compute_covariances(data, resp, counts, means)
        # Compute the loglikelihood at this iteration
        ll_latest = loglikelihood(data, weights, means, covariances)
        ll_trace.append(ll_latest)
        
        # Check for convergence in log-likelihood and store
        if (ll_latest - ll) < thresh and ll_latest > -np.inf:
            break
        ll = ll_latest
    
    if it % 5 != 0:
        print("Iteration %s" % it)
    
    out = {'weights': weights, 'means': means, 'covs': covariances, 'loglik': ll_trace, 
           'resp': resp}
    
    return out
```

## Fitting a Gaussian mixture model for the house data obtained from census.gov and zillow.com.

First, we load the SFrame data stored by running ipython notebook (tipping-point.ipynb).
The feature vectors are normalized ones with ['months_since', 'max_p', 'growth_rate', 'pop2010'], where 'pop2010' corresponds to population in 2010.

```python
house_data = gl.SFrame('data/census-zillow-sframe')
gl.canvas.set_target('ipynb')
import array
normalization_const = {}
normalization_const['months_since'] = np.sqrt(
    house_data['months_since'].apply(lambda x: x*x).sum())
normalization_const['max_p'] = np.sqrt(house_data['max_p'].apply(lambda x: x*x).sum())
normalization_const['growth_rate'] = np.sqrt(
    house_data['growth_rate'].apply(lambda x: x*x).sum())
normalization_const['pop2010'] = np.sqrt(house_data['pop2010'].apply(lambda x: x*x).sum())
house_data['months_since_norm'] = house_data['months_since']/normalization_const['months_since']
house_data['max_p_norm'] = house_data['max_p']/normalization_const['max_p']
house_data['growth_rate_norm'] = house_data['growth_rate']/normalization_const['growth_rate']
house_data['pop2010_norm'] = house_data['pop2010']/normalization_const['pop2010']

house_data['feature_vector'] = house_data.pack_columns(
    ['months_since_norm', 'max_p_norm', 'growth_rate_norm', 'pop2010_norm'])['X10']

house_data;
```

We need to come up with initial estimates for the mixture weights and component parameters. Let's take three feature vectors to be our initial cluster centers, and let's initialize the covariance matrix of each cluster to be diagonal with each element equal to the sample variance from the full data. We'll start by assuming each mixture component has equal weight. 

We will consider the number of clusters equal to 3.

```python
# Initalize parameters
num_clusters = 3
init_means = [house_data['feature_vector'][x] for x in np.random.choice(
        len(house_data), num_clusters, replace=False)]
# features = ['months_since', 'max_p', 'growth_rate', 'pop2010']

cov = np.diag([house_data['months_since_norm'].var(), house_data['max_p_norm'].var(), 
               house_data['growth_rate_norm'].var(), house_data['pop2010_norm'].var()])

init_covariances = [cov, cov, cov, cov]
init_weights = [1/4., 1/4., 1/4., 1/4.]

# Convert the feature vectors to numpy arrays
house_data_np = np.array([np.array(i) for i in house_data['feature_vector']])

# Run the EM algorithm on the house data using the above initializations. 
out = EM(house_data_np, init_means, init_covariances, init_weights, maxiter=20)
```

### Evaluating convergence

Let's start by plotting the log likelihood at each iteration - we know that the EM algorithm guarantees that the log likelihood can only increase (or stay the same) after each iteration, so if our implementation is correct then we should see an increasing function.

```python
ll = out['loglik']
plt.plot(range(len(ll)),ll,linewidth=4)
plt.xlabel('Iteration')
plt.ylabel('Log-likelihood')
plt.rcParams.update({'font.size':16})
plt.tight_layout()
```

### Evaluating uncertainty

Next we'll explore the evolution of cluster assignment and uncertainty. Remember that the EM algorithm represents uncertainty about the cluster assignment of each data point through the responsibility matrix. Rather than making a 'hard' assignment of each data point to a single cluster, the algorithm computes the responsibility of each cluster for each data point, where the responsibility corresponds to our certainty that the observation came from that cluster. 

We can track the evolution of the responsibilities across iterations to see how these 'soft' cluster assignments change as the algorithm fits the Gaussian mixture model to the data; one good way to do this is to plot the data and color each point according to its cluster responsibilities. Our data are four-dimensional, which can make visualization difficult, so to make things easier we will plot the data using only two dimensions, taking just the [months_since, max_p], [months_since, growth_rate], or other combinations among months_since, max_p, growth_rate, and pop2010.

```python
import colorsys
def plot_responsibilities_in_xy(df, feature_x, feature_y, resp, title):
    N, K = resp.shape
    #print N, K
    dict_name = {'months_since':'Months', 'max_p':'Median House Price ($)', 
                 'growth_rate':'Growth Rate (Number per Year)', 'pop2010':'Population in 2010'}
    
    HSV_tuples = [(x*1.0/K, 0.5, 0.9) for x in range(K)]
    RGB_tuples = map(lambda x: colorsys.hsv_to_rgb(*x), HSV_tuples)
    
    X = df[feature_x]
    Y = df[feature_y]
    resp_by_df_int = [[resp[n][k] for k in range(K)] for n in range(N)]
    cols = [tuple(np.dot(resp_by_df_int[n], np.array(RGB_tuples))) for n in range(N)]
    
    plt.figure()
    if feature_x == 'pop2010':
        for n in range(len(X)):
            plt.plot(X[n], Y[n], 'o', c=cols[n])
            plt.xscale('log')
    elif feature_y == 'pop2010':
        for n in range(len(Y)):
            plt.plot(X[n], Y[n], 'o', c=cols[n])
            plt.yscale('log')
    else:
        for n in range(len(X)):
            plt.plot(X[n], Y[n], 'o', c=cols[n])
    plt.title(title)
    plt.xlabel(dict_name[feature_x])
    plt.ylabel(dict_name[feature_y])
    plt.rcParams.update({'font.size':9})
    plt.tight_layout()
```

We now use the above plotting function to visualize the responsibilites after 1 iteration. 

```python
out = EM(house_data_np, init_means, init_covariances, init_weights, maxiter=1)
plot_responsibilities_in_xy(house_data, 'max_p', 'months_since', out['resp'], 
                            'After 1 iteration')
```

We now use the above plotting function to visualize the responsibilites after 20 iterations. We will see there are fewer unique colors; this indicates that there is more certainty that each point belongs to one of the four components in the model. 

```python
out = EM(house_data_np, init_means, init_covariances, init_weights, maxiter=20)
```

```python
import pylab 

features = ['months_since', 'max_p', 'growth_rate', 'pop2010']
for i,j in [(1,0), (2,0), (3,0),(1,2), (1,3), (2,3) ]:
    pij = 'p'+str(i)+str(j)
    pij = plot_responsibilities_in_xy(house_data, features[i], features[j], 
                                      out['resp'], 
                                      'After 20 iteration')
    fig_name = 'figures/fig_'+features[i]+'_'+features[j]
    pylab.savefig(fig_name)
```

The investable cities, i.e., high median house price and slow tipping point time, is found to typically belong to this class. This indicates that it is better not to invest housings in non-popular small cities, but large cities with high median house prices.
